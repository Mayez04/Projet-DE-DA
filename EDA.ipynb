```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette('husl')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
# Chargement des données
df = pd.read_csv('data/processed/combined_data.csv')
```

```python
# Vérification des valeurs manquantes
missing_values = df.isnull().sum()
missing_percentages = (missing_values / len(df)) * 100

print("\nPourcentage de valeurs manquantes par colonne :")
print("-" * 50)
for col, pct in missing_percentages.items():
    print(f"{col}: {pct:.2f}%")
```

```python
# 1. Suppression des lignes avec trop de valeurs manquantes
df_cleaned = df.dropna(thresh=len(df.columns) - 10)  # Garde les lignes avec au moins N-10 valeurs non-null
```

```python
df_cleaned.shape
```

```python
# 2. Remplacement des valeurs manquantes pour les colonnes numériques
numeric_columns = df_cleaned.select_dtypes(include=[np.number]).columns
for col in numeric_columns:
    df_cleaned[col] = df_cleaned[col].fillna(df_cleaned.groupby('region')[col].transform('median'))
```

```python
print("\nNombre de lignes avant nettoyage:", len(df))
print("Nombre de lignes après nettoyage:", len(df_cleaned))
```

```python
# Vérification des valeurs manquantes restantes
missing_after = df_cleaned.isnull().sum()
print("\nValeurs manquantes restantes :")
print(missing_after[missing_after > 0])
```

## HYPOTHÈSE 1 : Le niveau d’investissement dans l’éducation influence la réussite et la durée de la scolarisation.

### Dépenses en éducation (% du PIB) par continent

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='region', y='education_expenditure_gdp', ci=95)
plt.title('Dépenses en éducation par région (% du PIB)')
plt.xlabel('Région')
plt.ylabel('Dépenses en éducation (% du PIB)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Les régions montrent des différences significatives dans leurs investissements éducatifs :
- L'Europe investit le plus dans l'éducation en % du PIB
- L'Afrique et l'Asie présentent des investissements plus modérés
- Les écarts entre régions suggèrent des priorités budgétaires différentes

### 5 pays avec les dépenses les plus élevées en éducation par continent (% du PIB)

```python
# Group data by country and region to get mean education expenditure
df_mean = df.groupby(['country_code', 'region'])['education_expenditure_gdp'].mean().reset_index()

# Get top 5 countries for each region
top_5_by_region = df_mean.sort_values(by=['region', 'education_expenditure_gdp'], ascending=[True, False]) \
                        .groupby('region').head(5)

# Create the visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=top_5_by_region, 
            x='country_code',
            y='education_expenditure_gdp',
            hue='region',
            dodge=False)

plt.title("Top 5 pays par région - Dépenses en éducation (% du PIB)")
plt.xlabel("Pays")
plt.ylabel("Dépenses en éducation (% du PIB)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Certains petits pays investissent fortement dans l'éducation malgré des ressources limitées
- Les champions régionaux ne sont pas nécessairement les pays les plus riches
- On observe une grande variabilité intra-régionale dans les dépenses éducatives

### 5 pays avec les plus faibles dépenses en éducation par continent (% du PIB)

```python
# Group data by country and region to get mean education expenditure
df_mean = df.groupby(['country_code', 'region'])['education_expenditure_gdp'].mean().reset_index()

# Get bottom 5 countries for each region
bottom_5_by_region = df_mean.sort_values(by=['region', 'education_expenditure_gdp'], ascending=[True, True]) \
                        .groupby('region').head(5)

# Create the visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=bottom_5_by_region, 
            x='country_code',
            y='education_expenditure_gdp',
            hue='region',
            dodge=False)

plt.title("5 pays avec les plus faibles dépenses en éducation par région (% du PIB)")
plt.xlabel("Pays")
plt.ylabel("Dépenses en éducation (% du PIB)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Les pays avec les plus faibles investissements sont souvent confrontés à des défis économiques
- Certaines régions montrent une plus grande disparité dans les dépenses éducatives
- Les pays en conflit ou instabilité politique tendent à moins investir dans l'éducation

## HYPOTHÈSE 3 : Les inégalités de genre dans l’éducation varient selon le niveau d’enseignement, la région et le développement socio-économique des pays.

### meilleur et mauvaise parité par pays

```python
df['gender_gap_primary'] = abs(df['gender_ratio_primary'] - 1)
df['gender_gap_secondary'] = abs(df['gender_ratio_secondary'] - 1)
df['gender_gap_tertiary'] = abs(df['gender_ratio_tertiary'] - 1)

# Moyenne de l'écart pour chaque pays
df['gender_inequality_index'] = df[['gender_gap_primary', 'gender_gap_secondary', 'gender_gap_tertiary']].mean(axis=1)

# Moyenne par pays (au cas où on a plusieurs années)
df_mean = df.groupby(['country_code', 'region'])['gender_inequality_index'].mean().reset_index()

# Top 5 pays avec plus d'inégalités de genre par région
top_5_by_region = df_mean.sort_values(by=['region', 'gender_inequality_index'], ascending=[True, False]) \
                         .groupby('region').head(5)

print(top_5_by_region)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(
    data=top_5_by_region,
    x='gender_inequality_index',
    y='country_code',
    hue='region',
    dodge=False
)
plt.title("Top 5 pays par région avec les plus fortes inégalités de genre dans l'éducation")
plt.xlabel("Indice d'inégalité de genre (écart à la parité)")
plt.ylabel("Pays")
plt.tight_layout()
plt.show()
```

- Les inégalités de genre varient considérablement selon les régions
- Certains pays présentent des écarts très importants entre filles et garçons
- Les pays en développement sont plus touchés par ces inégalités
- Les situations de conflit amplifient souvent ces disparités

```python
df['gender_gap_primary'] = abs(df['gender_ratio_primary'] - 1)
df['gender_gap_secondary'] = abs(df['gender_ratio_secondary'] - 1)
df['gender_gap_tertiary'] = abs(df['gender_ratio_tertiary'] - 1)

# Moyenne des écarts pour obtenir un indice global d'inégalité
df['gender_inequality_index'] = df[['gender_gap_primary', 'gender_gap_secondary', 'gender_gap_tertiary']].mean(axis=1)

# Moyenne par pays (en cas de multiples années)
df_mean = df.groupby(['country_code', 'region'])['gender_inequality_index'].mean().reset_index()

# Top 5 pays les plus égalitaires par région (écart le plus faible à la parité)
top_5_equal_by_region = df_mean.sort_values(by=['region', 'gender_inequality_index'], ascending=[True, True]) \
                               .groupby('region').head(5)

print(top_5_equal_by_region)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(
    data=top_5_equal_by_region,
    x='gender_inequality_index',
    y='country_code',
    hue='region',
    dodge=False
)
plt.title("🏆 Top 5 pays par région avec la meilleure parité filles/garçons dans l'éducation")
plt.xlabel("Indice d'inégalité de genre (plus proche de 1 = meilleure égalité)")
plt.ylabel("Pays")
plt.tight_layout()
plt.show()
```

- Les pays développés tendent à avoir une meilleure parité
- L'Europe montre globalement de meilleurs ratios
- Certains pays en développement réussissent à atteindre une bonne parité
- Les politiques éducatives inclusives donnent des résultats positifs
