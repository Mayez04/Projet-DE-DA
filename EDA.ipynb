### D√©penses en √©ducation (% du PIB) par continent

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='region', y='education_expenditure_gdp', ci=95)
plt.title('D√©penses en √©ducation par r√©gion (% du PIB)')
plt.xlabel('R√©gion')
plt.ylabel('D√©penses en √©ducation (% du PIB)')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Les r√©gions montrent des diff√©rences significatives dans leurs investissements √©ducatifs :
- L'Europe investit le plus dans l'√©ducation en % du PIB

### 5 pays avec les d√©penses les plus √©lev√©es en √©ducation par continent (% du PIB)

```python
# Group data by country and region to get mean education expenditure
df_mean = df.groupby(['country_code', 'region'])['education_expenditure_gdp'].mean().reset_index()

# Get top 5 countries for each region
top_5_by_region = df_mean.sort_values(by=['region', 'education_expenditure_gdp'], ascending=[True, False]) \
                        .groupby('region').head(5)

# Create the visualization
plt.figure(figsize=(12, 6))
sns.barplot(data=top_5_by_region, 
            x='country_code',
            y='education_expenditure_gdp',
            hue='region',
            dodge=False)

plt.title("Top 5 pays par r√©gion - D√©penses en √©ducation (% du PIB)")
plt.xlabel("Pays")
plt.ylabel("D√©penses en √©ducation (% du PIB)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Certains petits pays investissent fortement dans l'√©ducation malgr√© des ressources limit√©es
- Les champions r√©gionaux ne sont pas n√©cessairement les pays les plus riches
- On observe une grande variabilit√© intra-r√©gionale dans les d√©penses √©ducatives

## HYPOTH√àSE 3 : Les in√©galit√©s de genre dans l‚Äô√©ducation varient selon le niveau d‚Äôenseignement, la r√©gion et le d√©veloppement socio-√©conomique des pays.

### meilleur et mauvaise parit√© par pays

```python
df['gender_gap_primary'] = abs(df['gender_ratio_primary'] - 1)
df['gender_gap_secondary'] = abs(df['gender_ratio_secondary'] - 1)
df['gender_gap_tertiary'] = abs(df['gender_ratio_tertiary'] - 1)

# Moyenne de l'√©cart pour chaque pays
df['gender_inequality_index'] = df[['gender_gap_primary', 'gender_gap_secondary', 'gender_gap_tertiary']].mean(axis=1)

# Moyenne par pays (au cas o√π on a plusieurs ann√©es)
df_mean = df.groupby(['country_code', 'region'])['gender_inequality_index'].mean().reset_index()

# Top 5 pays avec plus d'in√©galit√©s de genre par r√©gion
top_5_by_region = df_mean.sort_values(by=['region', 'gender_inequality_index'], ascending=[True, False]) \
                         .groupby('region').head(5)

print(top_5_by_region)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(
    data=top_5_by_region,
    x='gender_inequality_index',
    y='country_code',
    hue='region',
    dodge=False
)
plt.title("Top 5 pays par r√©gion avec les plus fortes in√©galit√©s de genre dans l'√©ducation")
plt.xlabel("Indice d'in√©galit√© de genre (√©cart √† la parit√©)")
plt.ylabel("Pays")
plt.tight_layout()
plt.show()
```

- Les in√©galit√©s de genre varient consid√©rablement selon les r√©gions
- Certains pays pr√©sentent des √©carts tr√®s importants entre filles et gar√ßons
- Les pays en d√©veloppement sont plus touch√©s par ces in√©galit√©s
- Les situations de conflit amplifient souvent ces disparit√©s

```python
df['gender_gap_primary'] = abs(df['gender_ratio_primary'] - 1)
df['gender_gap_secondary'] = abs(df['gender_ratio_secondary'] - 1)
df['gender_gap_tertiary'] = abs(df['gender_ratio_tertiary'] - 1)

# Moyenne des √©carts pour obtenir un indice global d'in√©galit√©
df['gender_inequality_index'] = df[['gender_gap_primary', 'gender_gap_secondary', 'gender_gap_tertiary']].mean(axis=1)

# Moyenne par pays (en cas de multiples ann√©es)
df_mean = df.groupby(['country_code', 'region'])['gender_inequality_index'].mean().reset_index()

# Top 5 pays les plus √©galitaires par r√©gion (√©cart le plus faible √† la parit√©)
top_5_equal_by_region = df_mean.sort_values(by=['region', 'gender_inequality_index'], ascending=[True, True]) \
                               .groupby('region').head(5)

print(top_5_equal_by_region)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(
    data=top_5_equal_by_region,
    x='gender_inequality_index',
    y='country_code',
    hue='region',
    dodge=False
)
plt.title("üèÜ Top 5 pays par r√©gion avec la meilleure parit√© filles/gar√ßons dans l'√©ducation")
plt.xlabel("Indice d'in√©galit√© de genre (plus proche de 1 = meilleure √©galit√©)")
plt.ylabel("Pays")
plt.tight_layout()
plt.show()
```

- Les pays d√©velopp√©s tendent √† avoir une meilleure parit√©
- L'Europe montre globalement de meilleurs ratios
- Certains pays en d√©veloppement r√©ussissent √† atteindre une bonne parit√©
- Les politiques √©ducatives inclusives donnent des r√©sultats positifs
